## You must compare all regions to see how they differ in the indicator

* Compare one year (pick the latest year) using bar chart, pie chartList item


```python
le1 =f1.loc[f1['indicator']=='SP.DYN.LE00.IN']
le1.head()
```

```python
life = le1.loc[le1['Year'] == 2020]
life
```

```python
f5 = life.groupby('Region')['value'].mean()
f5
```

```python
fig=px.bar(f5,x=f5.index,y='value', color=f5.index,title="2020 Life expectancy of 'Afghanistan', 'Japan', 'Philippines','St. Vincent and the Grenadines','Turkiye'")
fig.show()
```

```python
fig = px.pie(f5, values='value', names=f5.index, title="2020 Life expectancy of 'Afghanistan', 'Japan', 'Philippines','St. Vincent and the Grenadines','Turkiye'")
fig.show()
```

## Compare all Yers using Line Chart

```python
mean_le=le1.groupby(['Region','Year'])['value'].mean()
mean_le=mean_le.reset_index()
```

```python
fig=px.line(mean_le,x='Year',y='value',color='Region',title="'2020 Life expectancy of 'Afghanistan', 'Japan', 'Philippines','St. Vincent and the Grenadines','Turkiye'")
fig.show()
```

## Use all countries, and pick year 2020, make a scatter plot with x being GDP per Capita and y being Life Expentancy. This plot will tell you the relationship between wealth and health.

```python
f1=f1.loc[f1['Year']==2020]
f1.head()
```

```python
my_indicators = ["SP.DYN.LE00.IN", "NY.GDP.PCAP.PP.CD"]

df = f1[f1["indicator"].isin(my_indicators)]

df.shape
```

```python
gdp = pd.pivot_table(df, values = 'value', index='Country Name', columns = 'indicator').reset_index()
gdp
```

```python
px.scatter(gdp,x='NY.GDP.PCAP.PP.CD',y='SP.DYN.LE00.IN',color=gdp.index)
```
