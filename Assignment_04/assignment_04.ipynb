Definition of the above 9 variables:


**Variables**
1.   UNITID: 
Unit ID for institution, which is a unique number given to each postsecondary institutions surveyed by IPEDS.

2.   LINSTNM: Institution name reported in IPEDS

3.   CITY: Location of the Institute

4.   STABBR: State postcode of the Institute

5.   ZIP: ZIP code of the Institute

6.   ADM_RATE: Accreditor for institution

7.   UGDS: consists of the number of degree/certificate-seeking undergraduates enrolled in the fall as reported by the IPEDS

8.   TUITIONFEE_IN: Fee structrue for Bachelor's degree in Personal And Culinary Services.

9.   MN_EARN_WNE_P10: The average salary is calculated for all undergraduates at an institution receiving federal financial aid in a given academic year who are also working full-time but not in school.
